#### Gọi lại 2 models

```python
from model_classes import LinearRegression, ZScoreNormalizer
```

```python
model = joblib.load("models/linear_model.pkl")
normalizer = joblib.load("models/normalizer.pkl")
```

#### Tạo cột dự đoán

```python
feature_columns = [
    "Number_of_Customers_Per_Day",
    "Average_Order_Value",
    "Operating_Hours_Per_Day",
    "Number_of_Employees",
    "Marketing_Spend_Per_Day",
    "Location_Foot_Traffic"
]
```

#### Lấy giá trị trung bình của từng cột làm typical

```python
df = pd.read_csv("coffee_shop_revenue_cleaned.csv")
```

```python
mean_values = df.mean()

print(mean_values)
```

```python
new_cases = pd.DataFrame([
    [274, 6.3, 12, 8, 253, 535],  # Case 1: Typical
    [400, 6.3, 12, 8, 253, 535],  # Case 2: More Customers
    [275, 9.0, 12, 8, 253, 535],  # Case 3: Higher Order Value
    [275, 6.3, 12, 8, 450, 535]   # Case 4: Higher Marketing
], columns=feature_columns)
```

#### Chuẩn hóa

```python
X_new = new_cases[feature_columns]

X_new_scaled = normalizer.transform(X_new)

X_new_scaled
```

#### Dự đoán

```python
predictions = model.predict(X_new_scaled)

predictions
```

```python
new_cases["Predicted_Daily_Revenue"] = predictions

new_cases
```
